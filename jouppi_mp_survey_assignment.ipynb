### EXTRA: Comparing these responses to the current disenfranchisement laws in each state
*Note: I did this outside the 1.5hr time frame because I had time and was curious. Totally fine if you can't consider this part.*


**About the Data**

Data comes from the [Brennan Center](https://www.brennancenter.org/issues/ensure-every-american-can-vote/voting-rights-restoration/disenfranchisement-laws) map that divides the laws into 6 categories. I added a level # for severity that I based on reading the descriptions and the fact that the categories were color coded in diverging order. I would do more talking with the Brennan Center about whether that's a correct way to depict the data.

```python
# Reading in the voting law data. 51 records.
criminal_voting_laws_df = pd.read_csv('https://raw.githubusercontent.com/srjouppi/mp-survey-assignment/refs/heads/main/criminal-disenfranchisement-laws-state.csv')
display(criminal_voting_laws_df.head())
criminal_voting_laws_df.shape
```

```python
# Merging the survey responses with the voting law data. 21 records. Missing 1 because it dropped `0_ALL_STATES.`

vote_while_incarcerated_laws_df = vote_while_incarcerated_any_df.merge(criminal_voting_laws_df, left_on='state',right_on='state')
display(vote_while_incarcerated_laws_df.head())
vote_while_incarcerated_laws_df.shape
```

```python
# Which laws are represented in the dataset?
vote_while_incarcerated_laws_df.level.value_counts().sort_index()
```

```python
# To visualize this categorically, I'm going to make the `level` field a string.

vote_while_incarcerated_laws_df.level = vote_while_incarcerated_laws_df.level.astype(str)
vote_while_incarcerated_laws_df.head(1)
```

```python
custom_colors = {
    '1': '#90D6FF',
    '2': 'yellow',
    '4': 'orange',
    '5': 'red'
}

display(
    ggplot(vote_while_incarcerated_laws_df,
          aes('level','pct_of_respondents', size='count',fill='factor(level)'))
        + geom_point()
        + scale_fill_manual(values=custom_colors)
        + geom_hline(yintercept=0.556, color='black', linetype='dashed')
        + annotate('text', x=1.5, y=.53, label='All States = 55.6%', color='black', size=10)
        + ylim(0,1)
        + labs(
            y = "percent of respondents",
            x = "voter disenfranchisement level")

)
```

1. No disenfranchisement 🔵
2. Voting rights restored automatically upon release 🟡
3. Voting rights restored upon completion of sentence 🟠
4. Permanent disenfranchisement for some convictions 🔴
